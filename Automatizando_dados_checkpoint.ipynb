# Automatizando dados

```python
# Abrindo o arquivo com dados
import pandas as pd
dados = pd.read_csv('base_de_dados/wine_dataset.csv')
dados
```

```python
# passando os dados de string para int (tratamento de dados)

dados['style'] = dados['style'].replace('red'  , 0) #trocando o style red por 0
dados['style'] = dados['style'].replace('white', 1) #trocando o style white por 1

# atribuindo os valores para serem usados nos testes de machine learning

x = dados.drop('style', axis =1) # base de dados que levara em consideração os dados para se chegar ao valor da coluna y
y = dados['style']               # Coluna que recebera os dados da coluna de interesse

print(dados.shape)
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

#Criando um conjunto de dados para treino e para teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3) #Ao definir test_size 0.3 
# eu estou dizendo para o programa o que quero para o treino (30% nesse caso) e quanto eu quero para teste (70% neste exemplo)

print(x_treino.shape) #70% dos dados
print(y_treino.shape) #70% dos dados
print(x_teste.shape)  #30% dos dados
print(y_teste.shape)  #30% dos dados
```

```python
from sklearn.ensemble import ExtraTreesClassifier #Importo uma biblioteca inclusa dentro do ensemble que esta dentro do sklearn

# Criação do modelo
modelo = ExtraTreesClassifier() # atribuo a função Classificação extra de arvores para o modelo
modelo.fit(x_treino, y_treino)  # passo os dados de treino tanto de x quanto de y para o modelo a partir da função .fit()

# Imprimindo os resultados
resultado = modelo.score(x_teste, y_teste) # Acabei de automatizar o processo e estou procurando saber a sua eficiencia!
print('Acuracia: ', resultado)             # Eficiencia impressionante (Acurácia ou acerto eleado)!

# Esta célula é impressionante pois o programa, sozinho, procura advnhar qual a classe correspondente do vinho
# levando em consideração as caraceteristicas que o compõe dentro da base de dados :)
```
